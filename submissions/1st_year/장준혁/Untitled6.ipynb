```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 1. 한글 폰트(나눔고딕) 다운로드 및 즉시 적용 (세션 재시작 불필요)
font_url = 'https://github.com/google/fonts/raw/main/ofl/nanumgothic/NanumGothic-Regular.ttf'
font_name = 'NanumGothic-Regular.ttf'

if not os.path.exists(font_name):
    import urllib.request
    urllib.request.urlretrieve(font_url, font_name)

# 다운로드한 폰트를 matplotlib에 등록
fm.fontManager.addfont(font_name)
font_prop = fm.FontProperties(fname=font_name)
plt.rc('font', family=font_prop.get_name())
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# -------------------------------------------------------------

# 2. 데이터 불러오기 및 필터링 (2025년도 데이터만 & 과천시 제외)
file_name = '/workspaces/Daegun26Veritas/data/10224data.xlsx'
df = pd.read_excel(file_name)

# 2025년 데이터이면서 시군명이 과천시가 아닌 데이터만 필터링
df_filtered = df[(df['기준년도'] == 2025) & (df['시군명'] != '과천시')]

# 보기 좋게 1인당면적 기준 내림차순 정렬
df_sorted = df_filtered.sort_values(by='1인당면적', ascending=False)

# -------------------------------------------------------------

# 3. 그래프 그리기
plt.figure(figsize=(14, 8))

# 막대 그래프 그리기 (X축: 시군명, Y축: 1인당면적)
sns.barplot(data=df_sorted, x='시군명', y='1인당면적', palette='coolwarm')

# 그래프 제목 및 축 레이블 설정
plt.title('2025년 경기도 시군별 1인당 공원 면적 현황 (과천시 제외)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('시군명', fontsize=12, labelpad=10)
plt.ylabel('1인당 면적 (㎡)', fontsize=12, labelpad=10)

# X축 지역 이름들이 겹치지 않도록 45도 회전
plt.xticks(rotation=45)

# Y축 기준 격자선 추가
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 그래프 출력
plt.tight_layout()
plt.show()
```
